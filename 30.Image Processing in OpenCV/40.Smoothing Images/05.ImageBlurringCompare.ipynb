Image Blurring (Image Smoothing)
Image blurring is achieved by convolving the image with a low-pass filter kernel. It is useful for removing noise. It actually removes high frequency content (e.g: noise, edges) from the image resulting in edges being blurred when this is filter is applied. (Well, there are blurring techniques which do not blur edges). OpenCV provides mainly four types of blurring techniques.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('../../images/opencv_logo.png')
```

```python
filtered_average   = cv2.blur(img,(5,5))
filtered_gaussian  = cv2.GaussianBlur(img,(5,5),0)
filtered_median    = cv2.medianBlur(img,5)
filtered_bilateral = cv2.bilateralFilter(img,9,75,75)
```

```python
plt.subplot(231),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(filtered_average),plt.title('Average')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(filtered_gaussian),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(filtered_median),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(filtered_bilateral),plt.title('Bilateral')
plt.xticks([]), plt.yticks([])
plt.show()
```

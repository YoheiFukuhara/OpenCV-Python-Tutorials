```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('../../images/messi5.jpg',0)
```

```python
rows,cols = img.shape

#
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
print(M)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.gray()
plt.imshow(dst)
plt.show()
```

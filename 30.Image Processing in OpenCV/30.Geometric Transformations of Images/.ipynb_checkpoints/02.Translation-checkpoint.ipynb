```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('../../images/messi5.jpg',0)
```

```python
rows,cols = img.shape

#縦に100, 横に50ずらしている
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.gray()
plt.imshow(dst)
plt.show()
```

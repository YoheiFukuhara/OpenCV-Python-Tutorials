```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('../../images/gradient.png',0)
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)      #127以上の場合は255(白)、127以下は値変わらず
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)     #127以下の場合は0(黒)、127以上は値変わらず
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV) #上の逆

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
```

```python
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
